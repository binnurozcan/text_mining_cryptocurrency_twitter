```python
import pandas as pd
import snscrape.modules.twitter as sntwitter
from sqlalchemy.sql.expression import false
```

```python
litecoin = sntwitter.TwitterSearchScraper("litecoin")

tweets = []


for i, tweet in enumerate (litecoin.get_items()):
  if tweet.lang == 'en':
    data = [
        tweet.date, 
        tweet.id, 
        tweet.content, 
        tweet.user.username, 
        tweet.likeCount,
        tweet.retweetCount,
    ]
    tweets.append(data)
    if i > 10000:
      break

tweet_df = pd.DataFrame(
tweets, columns=["date", "id", "content", "username", "like_count", "retweet_count"])
tweet_df.to_csv("litecoin.csv", index =False)
```

Küçük harf dönüşümü. (Normalization)

```python

for data in kripto:
    data['content'] = data['content'].apply(lambda x: " ".join(x.lower() for x in x.split()))
     
bitcoin.head()
```

Noktalama işareti, emoji ve sayıların çıkarılması. (Removing punctuation)

```python
for data in kripto:
  data['content'] = data['content'].str.replace('[^A-Za-z0-9]+',' ', regex=True)
  data['content'] = data['content'].str.replace('\d','')
bitcoin.head()
```

```python
import re
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emotions
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags 
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
for data in kripto:
  data['content'] = data['content'].apply(lambda x: remove_emoji(x))

bitcoin.head()
```

Stopwords

```python

import nltk as nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

words = set(nltk.corpus.stopwords.words())
stop = stopwords.words('english')
newStopWords = ['oz','stopWord2']
stop.extend(newStopWords)
```

```python
for data in kripto:
  data['content'] = data['content'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
bitcoin.head()
```

Lemmatization

```python
nltk.download('omw-1.4')
from textblob import Word
nltk.download('wordnet')
for data in kripto:
  data['content'] = data['content'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
bitcoin.head()
```

Stemming

```python
from nltk.stem import PorterStemmer

st = PorterStemmer()
for data in kripto:
  data['content'] = data['content'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
bitcoin.head()
```

BITCOIN


Kelime bulutlari - Kelime Frekanslari

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Kelime Bulutu Oluşturma
text = ' '.join(bitcoin['content'].astype(str).tolist())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Kelime Bulutu - Bitcoin')
plt.show()

# Frekans Dağılımı Grafiği Oluşturma
word_counts = bitcoin['content'].str.split().explode().value_counts().sort_values(ascending=False).head(20)

plt.figure(figsize=(10, 5))
word_counts.plot(kind='bar')
plt.xlabel('Kelime')
plt.ylabel('Frekans')
plt.title('Kelime Frekansları - Bitcoin')
plt.show()

# Kelime listesini oluşturun
word_list = bitcoin['content'].str.split().tolist()

# İkili kelimeleri depolamak için boş bir liste oluşturun
bigrams = []

# Ardışık iki kelimeyi çiftler halinde kontrol edin
for words in word_list:
    for i in range(len(words) - 1):
        bigram = words[i] + ' ' + words[i + 1]
        bigrams.append(bigram)

# İkili kelime frekansını hesaplayın
bigram_counts = pd.Series(bigrams).value_counts().head(20)

# Frekans dağılımı grafiğini oluşturun
plt.figure(figsize=(10, 5))
bigram_counts.plot(kind='bar')
plt.xlabel('İkili Kelime')
plt.ylabel('Frekans')
plt.title('İkili Kelime Frekansları - Bitcoin')
plt.show()
```

duygu analizi


Genel Duygu Eğilimi:

```python
from textblob import TextBlob

# Bitcoin veri seti üzerinde genel duygu eğilimi analizi
bitcoin['sentiment'] = bitcoin['content'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Pozitif, negatif ve nötr duygulara sahip tweet sayıları
positive_tweets = bitcoin[bitcoin['sentiment'] > 0]
negative_tweets = bitcoin[bitcoin['sentiment'] < 0]
neutral_tweets = bitcoin[bitcoin['sentiment'] == 0]

print('Pozitif Tweet Sayısı:', len(positive_tweets))
print('Negatif Tweet Sayısı:', len(negative_tweets))
print('Nötr Tweet Sayısı:', len(neutral_tweets))

import matplotlib.pyplot as plt

# Pozitif, negatif ve nötr tweet sayıları
positive_count = len(positive_tweets)
negative_count = len(negative_tweets)
neutral_count = len(neutral_tweets)

# Grafik oluşturma
labels = ['Pozitif', 'Negatif', 'Nötr']
sizes = [positive_count, negative_count, neutral_count]
colors = ['green', 'red', 'blue']

plt.figure(figsize=(8, 5))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Genel Duygu Eğilimi - Bitcoin')
plt.axis('equal')
plt.show()
```
